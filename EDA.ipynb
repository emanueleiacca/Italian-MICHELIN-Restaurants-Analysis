```python
df = df_cleaned
columns_to_convert = ['HOTEL', 'AGRITURIMO', 'AFFITTACAMERE', 'CATERING', 'ALTRO']

# Replace 'X' with 1 in the specified columns
df[columns_to_convert] = df[columns_to_convert].replace('X', 1)

# If you want to convert all other values to 0, you can do this:
df[columns_to_convert] = df[columns_to_convert].fillna(0).astype(int)
df = pd.get_dummies(df, columns=['REGIONE'], drop_first=True)
```

```python
columns_to_convert = [
    'CAPITALE SOCIALE', 'DIPENDENTI\n2022', 'STELLE',
    'Indice di liquidità\n2020', 'Indice di liquidità\n2021', 'Indice di liquidità\n2022',
    'Indice di indip. finanz. (%)\n2020', 'Indice di indip. finanz. (%)\n2021', 'Indice di indip. finanz. (%)\n2022',
    'ROA (%)\n2020', 'ROA (%)\n2021', 'ROA (%)\n2022',
    'FOLLOWER\nCHEF\n2018', 'FOLLOWER\nCHEF\n2019', 'FOLLOWER\nCHEF\n2020',
    'FOLLOWER\nCHEF\n2021', 'FOLLOWER\nCHEF\n2022', 'FOLLOWER\nCHEF\n2023',
    'FOLLOWER\nRISTORANTE\n2018', 'FOLLOWER\nRISTORANTE\n2019', 'FOLLOWER\nRISTORANTE\n2020',
    'FOLLOWER\nRISTORANTE\n2021', 'FOLLOWER\nRISTORANTE\n2022', 'FOLLOWER\nRISTORANTE\n2023'
]

# Convert specified columns to numeric, handling errors by coercing them to NaN
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
from scipy.stats import zscore

z_scores = np.abs(zscore(df[numeric_columns]))
threshold = 3

outliers = (z_scores > threshold).sum(axis=1)
outliers_rows = df[outliers > 0]

print(f'Number of rows with outliers: {outliers_rows.shape[0]}')
print(outliers_rows)
```

```python
# Identify outliers using the IQR method
def find_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers

outliers_roa_2022 = find_outliers_iqr(df, 'ROA (%)\n2022')

print(f'Number of outliers in ROA (%) 2022: {len(outliers_roa_2022)}')
print(outliers_roa_2022[['ROA (%)\n2022']])
```

```python
from scipy.stats import zscore

df['ROA_zscore'] = zscore(df['ROA (%)\n2022'])
outliers_zscore = df[(df['ROA_zscore'] > 3) | (df['ROA_zscore'] < -3)]

print(f'Number of outliers detected by Z-score: {len(outliers_zscore)}')
print(outliers_zscore[['ROA (%)\n2022', 'ROA_zscore']])
```
